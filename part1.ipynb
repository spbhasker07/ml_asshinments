# Handle missing recourd by using sklern

```python
df=pd.read_csv('dataset/weather_data.txt') # dataset is folder name 
df.head()
```

```python
new_df=df.set_index('day')
new_df
```

```python
from sklearn.impute import SimpleImputer
```

```python
impute=SimpleImputer(missing_values=np.nan,strategy='mean')
impute
```

```python
new_df.iloc[:,0:2]=impute.fit_transform(new_df.iloc[:,0:2])
```

```python
new_df2=df.set_index('day')
```

```python
new_df2.iloc[:,0:2]=impute.fit_transform(new_df2.iloc[:,0:2])
```

```python
new_df3=df.set_index('day')
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
new_df3.iloc[:, :] = imputer.fit_transform(new_df3.iloc[:, :])
```

```python
new_df4=df.set_index('day')
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='constant',fill_value=4)
imputer
```

```python
new_df4.iloc[:,:]=imputer.fit_transform(new_df4.iloc[:,:])
new_df4
```
