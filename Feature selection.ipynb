```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd 
```

```python
iris=load_iris()
iris.keys()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
x=iris.data
y=iris.target
```

```python
y
```

# using scatter chart

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x[:,2],x[:,3],c=y)
plt.show()
```

```python
plt.scatter(x[:,0],x[:,3],c=y)
plt.show()
```

```python
plt.scatter(x[:50, 0], x[:50, 1], c='r', label='setosa')
plt.scatter(x[50:100, 0], x[50:100, 1], c='k', label='versicolor')
plt.scatter(x[100:150, 0], x[100:150, 1], c='b', label='virginica')
plt.legend()
plt.show()
```

# using corr()

```python
df=pd.DataFrame(x,columns=iris.feature_names)
df['target']=y
df.head()
df.corr()
```

#  using heatmap

```python
import seaborn as sb
```

```python
sb.heatmap(df.corr(),annot=True)
plt.show()
```
