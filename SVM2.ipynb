```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
iris=load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
sc=StandardScaler()
x_train_std=sc.fit_transform(x_train)
x_test_std=sc.transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

```python
svm=SVC(kernel='poly',random_state=0,gamma=10,C=10)
```

```python
svm.fit(x_train_std,y_train)
```

```python
y_pred=svm.predict(x_test_std)
```

```python
y_pred
```

```python
print('wrong prediction out of total ')
print((y_test!=y_pred).sum(),'/',((y_test==y_pred).sum()+(y_test!=y_pred).sum()))
```

```python
print('percentage accuracy',100*accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
c_m=confusion_matrix(y_test,y_pred)
c_m
```
