```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Student_Grades_Data .csv')
df.head()
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg=LinearRegression()
linreg.fit(x_train,y_train)
```

```python
print(f'a0={linreg.intercept_}')
print(f'a1={linreg.coef_}')
```

```python
y_predict=linreg.predict(x_test)
y_predict
```

```python
plt.scatter(x_train,y_train,color='#6B15E3')
plt.plot(x_train,linreg.predict(x_train),color='#1FEBBB')
plt.title('grade vs tittle')
plt.xlabel('time to study')
plt.ylabel('grades')
plt.show()
```

```python
predict_grad=linreg.predict([[10]])
predict_grad
```

```python
from sklearn import metrics
```

```python
accuracy = metrics.r2_score(y_test,y_predict)
print('accuracy',accuracy)
```

```python
print('mean squre error ',metrics.mean_squared_error(y_test,y_predict)
```
