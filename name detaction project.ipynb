```python
import random 
from nltk.corpus import names 
import nltk 
```

```python
def gender_features(word):
    return {'last_letter':word[-1]}
```

```python
labeled_names=([(name,'male') for name  in names.words('male.txt')]+ [(name, 'female') for name in names.words('female.txt')])
```

```python
labeled_names
```

```python
random.shuffle(labeled_names)
```

```python
labeled_names
```

```python
featuresets =[(gender_features(n),gender)
              for (n,gender) in labeled_names]
```

```python
featuresets
```

```python
train_set,test_set=featuresets[500:],featuresets[:500]
```

```python
len(train_set)
```

```python
len(test_set)
```

```python
classfire=nltk.NaiveBayesClassifier.train(train_set)
```

```python
# Predict new data
print(classfire.classify(gender_features('mahavir'))) 
```

```python
print(classfire.classify(gender_features('saumya'))) 
```

```python
while True:
    name=input("Name: ")
    print(classfire.classify(gender_features(name)))
    ch=input("Do U continue this Loop(y/n):")
    if(ch=='y' or ch=='Y'):
        continue
    else:
        print("Thanks")
        break
```

```python
# it shows accurancy of our classifier
print(nltk.classify.accuracy(classfire, train_set)) 
```
