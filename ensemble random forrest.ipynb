```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Position_Salaries.csv')
df.head()
```

```python
x=df.iloc[:,1:2].values
y=df.iloc[:,2].values
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg=RandomForestRegressor(n_estimators=10,random_state=5)
reg.fit(x,y)
```

```python
y_pred=reg.predict([[7]])
y_pred
```

```python
pred=reg.predict(x)
pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,pred)
```
