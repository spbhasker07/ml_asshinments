```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
x=iris.data
```

```python
x
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(x)
df.columns=iris.feature_names
df['target']=iris.target
df
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
km.fit(x)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
df[' k means predicated label']=km.labels_
```

```python
df
```

```python
pd.crosstab(iris.target,km.labels_)
```

```python
km.inertia_
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,random_state=0)
    km.fit(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss,marker='o')
plt.xlabel('number of cluster')
plt.ylabel('wcss')
plt.show()
```
