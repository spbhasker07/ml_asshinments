```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Height_Weight_Dataset.csv')
df.head()
```

```python
x=df.iloc[:,0:1].values
y=df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg=LinearRegression()
linreg.fit(train_x,train_y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynom=PolynomialFeatures(degree=6)
x_polynom=polynom.fit_transform(train_x)
x_polynom
```

```python
poly_reg=LinearRegression()
poly_reg.fit(x_polynom,train_y)
```

```python
plt.scatter(train_x,train_y,color='red')
plt.plot(train_x,poly_reg.predict(polynom.fit_transform(train_x)),color='blue')
plt.title('polynom regration')
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```
