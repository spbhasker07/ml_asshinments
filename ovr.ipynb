```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(multi_class='ovr')
model.fit(x,y)
```

```python
sample=[1.5,.8,3.5,.9]
print(model.predict_proba([sample]))
print(model.predict([sample]))
```

```python
model.coef_
```

```python
model.intercept_
```

```python
z0=-0.44565052*1.5+0.90054581*.8+-2.32023343*3.5+-0.98112874*.9+6.68931554
z1=-0.17917535*1.5+-2.12878487*.8+0.69671839*3.5+ -1.27500517*.9+5.58592907
z2=-0.39270716*1.5+-0.51442952*.8+ 2.93064063*3.5+2.41608802*.9+-14.43598319
```

```python
print(z0,z1,z2)
```

```python
model.decision_function([sample])
```

```python
import numpy as np
```

```python
p0=1/(1+np.exp(-z0))
p1=1/(1+np.exp(-z1))
p2=1/(1+np.exp(-z2))
```

```python
print(p0,p1,p2)
```

```python
model.predict_proba([sample])
```

```python
pn0=p0/(p0+p1+p2)
pn1=p1/(p0+p1+p2)
pn2=p2/(p0+p1+p2)
print(pn0,pn1,pn2)
```

```python
print(model.predict([sample]))
```

```python

model = LogisticRegression(multi_class='multinomial', max_iter=200)
model.fit(x, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
sample=[1.5,.8,3.5,.9]
z0=-0.4248362*1.5+ 0.96741615*.8+-2.51536487*3.5+-1.08235714*.9+9.85393186
z1=0.53512448*1.5+-0.32086871*.8+-0.20703587*3.5+-0.94315282*.9+ 2.23224793
z2=-0.11028828*1.5+-0.64654745*.8+ 2.72240073*3.5+2.02550996*.9+-12.08617979
print(z0,z1,z2)
```

```python
model.decision_function([sample])
```

```python
p0=np.exp(z0)/(np.exp(z0)+np.exp(z1)+np.exp(z2))
p1=np.exp(z1)/(np.exp(z0)+np.exp(z1)+np.exp(z2))
p2=np.exp(z2)/(np.exp(z0)+np.exp(z1)+np.exp(z2))
print(p0,p1,p2)
```
